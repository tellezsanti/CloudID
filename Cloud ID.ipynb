```python
import os
import numpy as np
import pandas as pd
import random

import matplotlib.pyplot as plt

from PIL import Image, ImageFilter

from utils import DataGenerator

from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten, MaxPooling2D, BatchNormalization, Dropout
from keras.utils import to_categorical
from keras.callbacks import ModelCheckpoint
from keras.models import load_model

from coremltools.converters.keras import convert
```

```python
# visualize random clouds from dataset

cloud_class = np.random.choice(os.listdir('data'))
cloud_file_name = np.random.choice(os.listdir(os.path.join('data/', cloud_class)))

print(cloud_class)
Image.open(os.path.join('data', cloud_class, cloud_file_name))
```

```python
# split training, validation and test

train_size = 0.7
validation_size = 0.15

file_names = []
cloud_classes = os.listdir('data')

for cloud_class in cloud_classes:
    file_names = file_names + os.listdir(os.path.join('data', cloud_class))

# random order for cloud types
random.shuffle(file_names)

first_split = int(train_size * len(file_names))
second_split = int((train_size + validation_size) * len(file_names))

train_files = file_names[:first_split]
validation_files = file_names[first_split:second_split]
test_files = file_names[second_split:]
```

```python
# initialize training parameters

input_size = (128, 128, 3)
n_classes = len(cloud_classes)

params = {'dim': input_size[:2],
          'batch_size': 16,
          'n_classes': n_classes,
          'n_channels': input_size[2],
          'shuffle': True}

epochs = 500
lr = 1e-5
batch_size = 16
```

```python
# build data generators

partition = {}

partition['train'] = train_files
partition['validation'] = validation_files
partition['test'] = test_files

labels = {}
for file in file_names:
    labels[file] = file[:2]

class_encodings = {}
for i, cloud_class in enumerate(cloud_classes):
    class_encodings[cloud_class] = i

encoded_labels = {}
for img_id in list(labels.keys()):
    encoded_labels[img_id] = class_encodings[labels[img_id]]
    
# Generators
training_generator = DataGenerator(partition['train'], labels, encoded_labels, **params)
validation_generator = DataGenerator(partition['validation'], labels, encoded_labels, **params)
test_generator = DataGenerator(partition['test'], labels, encoded_labels, **params)
```

```python
# keras sequential model

model = Sequential()

model.add(Conv2D(32, (3, 3), activation = 'relu', input_shape = input_size))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.1))

model.add(Conv2D(32, (5, 5), activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.2))

model.add(Conv2D(64, (3, 3), activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.3))

model.add(Conv2D(32, (3, 3), activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.3))

model.add(Conv2D(32, (3, 3), activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.3))

model.add(Flatten())
model.add(Dense(64, activation = 'relu'))
model.add(Dense(n_classes, activation = 'softmax'))

model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
model.summary()
```

```python
# train model

save_best = ModelCheckpoint(
    filepath='cloud_best.h5',
    save_weights_only=False,
    monitor='val_accuracy',
    mode='max',
    save_best_only=True)

history = model.fit_generator(generator = training_generator,
                              validation_data = validation_generator,
                              epochs = epochs, 
                              callbacks = [save_best],
                              use_multiprocessing = True,
                              workers = 4);
```

```python
model = None
```

```python
plt.plot(np.array(history.history['accuracy']), label = 'accuracy')
plt.plot(np.array(history.history['val_accuracy']), label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc = 'lower right')
```

```python
model = load_model('cloud_best.h5')
test_loss, test_acc = model.evaluate_generator(test_generator, verbose = 2)
print("Test set loss: " + str(np.round(test_loss, 3)))
print("Test set accuracy: " + str(np.round(test_acc, 3)))
```
